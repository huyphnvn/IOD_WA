```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vector multiplication
v1 = np.array([1, 10, 12])
v2 = np.array([2, -4, 0])
# dot product
print(np.dot(v1, v2))
```

```python
# Plotting the vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim3d([0, 3])
ax.set_xlabel('X')
ax.set_ylim3d([-38, 12])
ax.set_ylabel('Y')
ax.set_zlim3d([0, 12])
ax.set_zlabel('Z')
# Vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='v2')
# Dot product vector
dot_product_vector = v1 * v2
ax.quiver(0, 0, 0, dot_product_vector[0], dot_product_vector[1], dot_product_vector[2], color='g', label='Dot Product', pivot="tail")
```
