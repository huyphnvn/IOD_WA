### Check Column Names

Check column names and clean by converting all names to lowercase, replacing spaces with underscores (`_`) and removing question marks or asterisk (`*`) characters.

```python
# ANSWER
rock_data.columns = rock_data.columns.str.replace(' ', '_')
rock_data.columns = rock_data.columns.str.replace('*', '')
rock_data.columns = rock_data.columns.str.replace('?', '')
rock_data.columns = rock_data.columns.str.lower()
```

### Write a function that converts a column in a DataFrame to a numeric type and otherwise replaces entries with np.nan

```python
# ANSWER
def convert_column(df, column):
    df[column] = pd.to_numeric(df[column], errors='coerce')
    return df[column]
```

```python
def convert_column_only(column):
    column = pd.to_numeric(column, errors='coerce')
    return column
```

### Apply this last function to your dataset

```python
# ANSWER
convert_column(df=rock_data, column='release_year')
```

```python
for col in rock_data.columns:
    convert_column(df=rock_data, column=col)
```

```python
rock_data
```

```python
new_rock_data = rock_data.apply(convert_column_only, axis=0)
new_rock_data
```
